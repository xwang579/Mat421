```python
from numpy.linalg import det

M = np.array([[3,2,-4,3], 
             [3,0,11,-1], 
             [1,-3,0,3],
             [0,-2,0,1]])
print('M:\n', M)

I = np.eye(4)
print('I:\n', I)
print('M*I:\n', np.dot(M, I))

from numpy.linalg import \
             cond, matrix_rank

print('Determinant:\n %.1f'%det(M))

print('Condition number:\n', cond(M))
print('Rank:\n', matrix_rank(M))
```

```python
import numpy as np
from numpy.linalg import qr
```

```python
a = np.array([[1, 7], 
              [-2, 4]])

q, r = qr(a)

print('Q:', q)
print('R:', r)

b = np.dot(q, r)
print('QR:', b)
```

```python
import numpy as np
from numpy.linalg import eig

a = np.array([[1, 7], 
              [-2, 4]])
w,v=eig(a)
print('E-value:', w)
print('E-vector', v)
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
# generate x and y
x = np.linspace(0, 3, 130)
y = 5 + 2*x + x * np.random.random(len(x))
# assemble matrix A
A = np.vstack([x, np.ones(len(x))]).T

# turn y into a column vector
y = y[:, np.newaxis]
```

```python
# Direct least square regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)
```

```python
# plot the results
plt.figure(figsize = (15,10))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_d = np.array([0, 3, 5, 7, 9, 11, 13, 15, 17])
y_d = np.array([0, 0.8, 0.9, 0.1, -0.6, -0.8, -1, -0.9, -0.4])

plt.figure(figsize = (15, 9))
for i in range(2, 8):
    
    # get the polynomial coefficients
    y_est = np.polyfit(x_d, y_d, i)
    plt.subplot(2,5,i)
    plt.plot(x_d, y_d, 'o')
    # evaluate the values for a polynomial
    plt.plot(x_d, np.polyval(y_est, x_d))
    plt.title(f'Polynomial order {i}')

plt.tight_layout()
plt.show()
```
