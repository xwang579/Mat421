Law of cosines
If Œ∏ is an angle between two vectors u and v in ‚Ñù2 or ‚Ñù3 , then the law of cosines says that

|u-v|^2 == u^2 + v^2 - 2 |u||v| cos(Œ∏)

This law can be used to determine the angle between two vectors.

Find the angle between the vectors ùë£=[x,y,z]and ùë§=[a,b,c]. Where x,y,z,a,b,c can be any arbitary number.

```python
# for example let ùë£=[-3,7,21]and ùë§=[4,-2,9]

from numpy import arccos, dot

v = np.array([[-3,7,21]])
w = np.array([[4,-2,9]])
theta = \
    arccos(dot(v, w.T)/(norm(v)*norm(w)))
print(theta)
# theta is the angle between vectoe u and v
```

```python
#Calculating the Determinant
#all the matrix must be square (have the same number of rows as columns). Like 2x2,3x3,etc...

from numpy.linalg import det

M = np.array([[2,4,1,-2], 
             [2,2,-8,0], 
             [1,2,-1,8],
             [0,2,-2,0]])
print('M:\n', M)

print('Determinant: %.1f'%det(M))
I = np.eye(4)
print('I:\n', I)
print('M*I:\n', np.dot(M, I))
```

```python
#The inverse of a square matrix ùëÄ is a matrix of the same size, ùëÅ . ùëÄ‚ãÖùëÅ=ùêº

from numpy.linalg import inv

print('Inv M:\n', inv(M))

M = np.array([[2,4,1,-2], 
             [2,2,-8,0], 
             [1,2,-1,8],
             [0,2,-2,0]])

print('det(M):\n', det(M))
```

```python
from numpy.linalg import \
             cond, matrix_rank
# A matrix is called full rank. if rank (ùê¥)=min(ùëö,ùëõ)
print('Condition number:\n', cond(M))
print('Rank:\n', matrix_rank(M))
y = np.array([[1], [2], [1],[2]])
M_y = np.concatenate((M, y), axis = 1)
print('Augmented matrix:\n', M_y)
```
