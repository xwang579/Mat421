Law of cosines
If Œ∏ is an angle between two vectors u and v in ‚Ñù2 or ‚Ñù3 , then the law of cosines says that

|u-v|^2 == u^2 + v^2 - 2 |u||v| cos(Œ∏)

where |w| denotes the Euclidean norm of a vector w.
This law can be used to determine the angle between two vectors.
This is where the formula from.

Find the angle between the vectors ùë£=[x,y,z]and ùë§=[a,b,c]. Where x,y,z,a,b,c can be any arbitary number.

```python
# for example let ùë£=[-3,7,21]and ùë§=[4,-2,9]

from numpy import arccos, dot

v = np.array([[-3,7,21]])
w = np.array([[4,-2,9]])

theta = \
    arccos(dot(v, w.T)/(norm(v)*norm(w)))
print(theta)
# theta is the angle between vectoe u and v
```

```python
#Calculating the Determinant
#all the matrix must be square (have the same number of rows as columns). Like 2x2,3x3,etc...

from numpy.linalg import det

M = np.array([[2,4,1,-2], 
             [2,2,-8,0], 
             [1,2,-1,8],
             [0,2,-2,0]])
print('M:\n', M)

print('Determinant: %.1f'%det(M))
I = np.eye(4)
print('I:\n', I)
print('M*I:\n', np.dot(M, I))
```

```python
#The inverse of a square matrix ùëÄ is a matrix of the same size, ùëÅ . ùëÄ‚ãÖùëÅ=ùêº

from numpy.linalg import inv

print('Inv M:\n', inv(M))

M = np.array([[2,4,1,-2], 
             [2,2,-8,0], 
             [1,2,-1,8],
             [0,2,-2,0]])

print('det(M):\n', det(M))
```

```python
from numpy.linalg import \
             cond, matrix_rank

# A matrix is called full rank. if rank (ùê¥)=min(ùëö,ùëõ)

print('Condition number:\n', cond(M))
print('Rank:\n', matrix_rank(M))
y = np.array([[1], [2], [1],[2]])
M_y = np.concatenate((M, y), axis = 1)
print('Augmented matrix:\n', M_y)
```

Gauss-Seidel Method

With the Gauss-Seidel method, we use the new values x_i^(k+1) as soon as they are known. Once we have computed x_1^(k+1) from the first equation, its value is then used in the second equation to obtain the new x_2^(k+1).
Where in the kth iteration remain unchanged until the entire (k+1)th iteration has been calculated.



```python
x1 = 0
x2 = 0
x3 = 0
epsilon = 0.01
converged = False

x_old = np.array([x1, x2, x3])

print('Iteration results')
print(' K:   x1,    x2,    x3 ')
for k in range(1, 50):
    x1 = (4-3*x2+3*x3)/8
    x2 = (25+2*x1-2*x3)/(-6)
    x3 = (-18-3.3*x1-5*x2)/(-3)
    x = np.array([x1, x2, x3])
    # check if it is smaller than threshold
    dx = np.sqrt(np.dot(x-x_old, x-x_old))
    
    print("%d: %.4f, %.4f, %.4f"%(k, x1, x2, x3))
    if dx < epsilon:
        converged = True
        print('Converged!')
        break
        
    x_old = x

if not converged:
    print('Not converge, increase the # of iterations')
```

**Use python to solve the systems of equations.**




```python
import numpy as np

A = np.array([[8, 4, 3], 
              [2, 6, 9], 
              [ 1, -5,0]])
y = np.array([-4,5, 7])

x = np.linalg.solve(A, y)
print(x)
```

```python
A_inv = np.linalg.inv(A)

x = np.dot(A_inv, y)
print(x)
```

```python
#use the matrix A above to get L and U
from scipy.linalg import lu

P, L, U = lu(A)
print('P:\n', P)
print('L:\n', L)
print('U:\n', U)
print('LU:\n',np.dot(L, U)) #permutation matrix on ùê¥
```
